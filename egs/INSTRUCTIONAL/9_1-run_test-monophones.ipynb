# 9.1: `run_test.sh` - Decoding the `test` set using `monophones` acoustic model

Our script `run_test.sh` really just calls one script: `decode.sh` which "does all the dirty work".  If you'd like to get some information about each of the individual `C` commands used in `decode.sh`, have a look at [this blog post](http://jrmeyer.github.io/asr/2016/09/12/Using-built-GMM-model-Kaldi.html).

To start with, we will decode using our `monophone` acoustic model.

## Setting `kaldi_config.json`

Edit the **last** section (`run_test`) of `kaldi_config.json` to use the following arguments:
 - `graph_dir`=`exp/monophones/graph`  # will use the HCLG from the monophones acoustic model
 - `data_test_dir`=`data/test_dir`     # location of already-processed (during run_prepare_data.sh) test data
 - `decode_dir`=`exp/monophones/decode_test_dir`         # location to house all logs and output files
 - `num_processors`=`4`  # this will parallelize using **four** cores
 - `weight`=`10`         # default value for language modeling weight
 - `beam`=`6.0`          # default value for beam search 
 - `max_active`=`7000`   # default value for max_active nodes
 - `save_to`=`exp/experiments/default_experiment_monophones`   # location to save results and predicted transcripts
 - `non_vanilla_lda_mllt_align_hyperparameters`=`null`

````bash
head -n22 run_test.sh
````

Run `run_test.sh`.  If you set `n=250` in the first notebook, this step could take up to 30 minutes.

You will notice a few things in the output above (working our way through the output in reverse order):

 1. `WER` v `SER`
   - `word error rate` is the percentage of words transcribed **incorrectly**.  `sentence error rate` is the percentage of **audio samples** with **at least one** error in them.
 2. `{ws}er_*`
   - You see **five** different `word error rate`s, each with a different `integer` after the `_`.  These are **five** different decodings, using **five** different weights for the language model.  We'll look at this in more detail later.
 3. `time to decode and score` *only* includes the time for decoding (this will become important next week when we alter hyperparameters)
 4. a script called `analyze_lat.sh` was run multiple times, each time with the different language model weight.  
 
We will look at the output of this script and all the other files generated by `run_test.sh` in a later notebook.
